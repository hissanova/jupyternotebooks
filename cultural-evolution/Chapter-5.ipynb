## 5. 文化大進化の数理

### 5.1 文化大進化とは何か
文化大進化(Cultural macroevolution)とは、集団レベルでの文化進化の事。
- 集団間の文化伝達によって起こる文化進化。
- 一方、文化小進化(cultural microevolution)は、 個人レベル、すなわち、個体間の文化伝達によって起こる文化進化のことである。
- 集団を単位として扱った方が便利な場合がある
- 集団として扱う以外にない場合がある
	- 例
    	- 遺跡から発掘された考古遺物。
        - Ethnographic atlas: https://d-place.org/contributions/EA
        	- 各集団に見られる様々な文化要素が記載されている＝複数の個人の集合としての集団に対して、ある属性があるかないかの記録
        - 言語学：地域・集団ごとに話されている言語の多様性。
        	- どの程度の差があるか。どんな原因でその差が生じたのか。
        - 文化心理学：集団間の多様性が関心。「西洋」や「東洋」といった大きな区分が扱われることもある。
        - 考古学：地域、遺跡、文化層のような単位が扱われる。「縄文文化」、「弥生文化」といったまとまりも対象になる。


**目的**：現在および過去の文化多様性を記述し、どのようにしてその多様性が形成されてきたかを明らかにする。形成プロセスそれ自体の多様性と共通性を知る。

### 5.2 分岐・絶滅・文化伝搬
- 文化伝達
	- 垂直伝達(vertical transmission, phylogenesis)
    	- 歴史を共有することによる文化（形質）の共有。一つの集団が二つに分岐することで起こる。
    - 水平伝達(horizontal transmission, ethnogenesis)
    	- 集団間の文化伝達による、文化（形質）の共有。
- 適応(adaptation)
	- イノベーション(innovation)
    	- 集団が新しい文化形質を獲得すること。
        - 異なる集団が、類似した環境にいる場合、非常に似た文化形質を獲得することがある。
- 絶滅(extinction)
	- 構成員が全滅する、またはある文化が完全に失われる事。

何が文化大進化の主要な原因であるかは以前から議論されているが、文化進化研究が、垂直伝達、水平伝達、環境適応の相対的な寄与を定量的に捉える視点を提供。
#### 5.2.1 大進化における垂直伝達と水平伝達
文化大進化のモデルの一例： Charles Nunn の空間的モデル。

モデル概要
- 2次元格子モデル。
- 各集団は、2つの連続値形質 $X,Y$ をもつ。
- Algorithm
	- Step 1 (Extinction)
    	- 確率 $P_{extinction}$ で絶滅する. 残りの集団が１つだけの場合は $P_{extinction}=0$
        - 生き延びた場合、隣接する空いているセルに入植可能。
        - 絶滅した場合、全員死亡。
    - Step 2 (Immigration/Bifurcation)
    	- 入植可能なセルは東西南北にあるセル。
        - 入植後は異なる集団として扱う。（分岐＝垂直伝達のプロセス）
        - 入植確率 $P_{immigration}=0.96$ で固定。
    - Step 3 (Horizontal transmission)
    	- 水平伝達の起こる確率 $P_{horizontal}$ 。
        - 伝達先の社会の形質 $(X,Y)$ は、伝達元の社会の形質 $(X',Y')$ で置き換えられる。
        - 水平伝達はすべての社会で同時に行われるとする。（結果的に、文化形質の交換が起こる時がある）
        - ある社会に水平伝達する社会は、離接する社会からランダムに１つ選ぶ。
    - Step 4 (Evolution of traits)
    	- $(X, Y)$ をブラウン運動モデルで進化させる。
- Step 1~4 までを1世代とし、60世代繰り返す
- 初期値は、左端列、中ほどの行に、一つの集団

実験概要
- 60世代後の2次元格子上の形質値 $(X,Y)$ の分布を調べた。
- パラメーターを変化させることで、形質の類似度と地理的距離、系統的距離（分岐してからの時間）がどう影響されるかを調べた。
- 分析結果
	- 水平伝達の確率が上昇するほど
    	- 形質の類似度と地理的距離との相関は強くなった
    	- 形質と系統的距離との相関は弱くなった
    - 統計解析の結果から、そのデータを生成したプロセスを判別できるのかを検討
    	- 相関係数ベース（マンテル検定）：垂直伝達と水平伝達を区別することは難しいと報告

Crema et al は離散値（2値？）形質を扱えるようモデルを拡張した。


#### 5.2.2 騎馬技術の拡散のモデリング
Peter Turchin et al による、軍事技術の拡散に伴う巨大な政治体（e.g. 帝国）の出現のモデル化。

モデル概要
- 対象：ユーラシア大陸とアフリカ大陸
- 100km×100km グリッドに区切られている。
    - 各グリッドには
    	- 農耕の有無、バイオーム、標高、によって特徴づけられる。
    	- ある1つの集団が居住している。
- 各集団は
	- 何らかの政治体に属す（本には陽に書いてない）
    - 以下の2値形質をもつ
        - 社会的形質 $a =(a_1,\dots,a_{n_a})\in \{0,1\}^{n_a}$ 
            - $\forall i, P(a_i = 0 \mapsto 1) < P(a_i = 1 \mapsto 0) $
        - 農耕技術 $\tau_{a} \in \{0,1\}$
            - $\tau_{a} = 1$ の場合のみ、隣接集団に攻撃可能。
        - 軍事技術 $\tau_{m}=(\tau_1,\dots,\tau_{n_m}) \in \{0,1\}^{n_m}$
            - $|\tau_m|_1$ の値が大きいほど、負かした集団の文化が消滅し、勝った方の文化を受け入れる確率が上がる（文化の根絶）。
            
- 戦争のルール
	- 戦争の始め方
    	- ある政治体の境界領域から無作為にグリッド $g_{i,j}$ を選ぶ。
        - $g_{i,j}$ に隣接したグリッドから無作為に選び、そのグリッドが別の政治体だった場合、確率 $P_{war}$ で戦争が始まる。
        - すべての境界上のグリッドは毎年一度、攻撃の機会を得る。
        - 意思決定の順番は毎回ランダム。
        - 沿岸のグリッドも攻撃可能。
        - 沿岸のグリッドの攻撃先に海が選ばれた場合、距離 $d_{sea}$ 内にある沿岸部のグリッドに通常通り攻撃する。
    - 勝敗の決定要因
    	- 言葉で
	    	- 政治体の集団サイズ＝グリッドの数が多いと勝ちやすい
    	    - 社会的形質の保持数 $|a|$が多いと勝ちやすい
        	- 標高が高いと負けにくい
        - 数式で
        	- 攻撃側の戦闘能力 $P_{att}:= 1 + \beta_A \bar{A}_{att}N_{att}$ 
                -  $ \bar{A}_{att} = \frac{\sum_i \sum_k a_{i,k}}{N_{att}}$、$a_{i,k} \in \{0,1\}$ は注目している政治体の $i$ 番目のコミュニティの $k$ 番目の社会性形質。
                - $N_{att}$ は注目している政治体に属するコミュニティの数。
                - $\beta_A$ 変換パラメータ
            - 防御側の戦闘能力 $P_{def}:= 1 + \beta_A \bar{A}_{def}N_{def} + \gamma_{war} E_{def}$
            	- $E_{def}$ ：防御集団の住むグリッドの標高
                - $\gamma_{war}$ 変換パラメータ
            - 攻撃が成功する確率 $P_{success}:=\frac{P_{att} - P_{def}}{P_{att}+P_{def}}$
            	- $P_{success} <0$ の場合、攻撃は必ず失敗する。
    - 戦争の結果（併合と文化の根絶）
    	- 負けた集団は、勝った政治体に併合される。
        - 軍事技術形質の合計が高いほど文化の根絶が起こりやすい
        	- 文化根絶が起こる確率 $P_{ethno}:=\epsilon_{min} + (\epsilon_{max} - \epsilon_{min})\bar{M}_{att} - \gamma_{ethno}E_{def}$
            - $\bar{M}_{att}$ は、攻撃側の軍事技術形質の平均。よって、$0\le \bar{M}_{att}\le 1$.  ($\bar{M}_{att}:=\frac{|\tau_m|}{n_m}$?)
            - $\gamma_{ethno}$ は、標高が文化根絶を妨げる効果をコントロールするパラメータ。
            - $P_{ethno} <0$ の場合、文化根絶は起こらない。
        	- 文化の根絶が起こると、負けた集団の社会的形質の配列が勝った集団の社会的形質の配列に置き換えられる。
    - 軍事技術の伝達
    	- 水平伝達で行われる。
        - 各タイムステップ毎に、すべての農地の中からランダムに一つの農耕地を選び、その集団の持っている軍事技術 $\tau_m$ から、ランダムに成分 $\tau_{m,i}$ を一つ選ぶ
        	- $\tau_{m,i} = 0$ なら、何も起こらない
            - $\tau_{m,i} = 1$ の場合、隣接４グリッドのうち一つを選ぶ（その軍事技術を$\tau^{\prime}_{m}$ とする）。もし、$\tau^{\prime}_{m,i} = 0$ ならば、確率 $\sigma$ で、$\tau^{\prime}_{m,i} =0 \mapsto 1$ になる。
            - いったん伝搬された軍事技術は失われないとする。
- 社会性形質の獲得
	- 突然変異か、文化同化（戦争による文化根絶）によって獲得される。
    	- 突然変異の起こる確率
        	- $ P(a_{i} = 0 \mapsto 1) = \mu_{01}$
            - $ P(a_{i} = 1 \mapsto 0) = \mu_{10}$
            - $ \mu_{01} \ll \mu_{10}$ とする（社会的性質は失われやすい）。
            - したがって、戦争がない場合の社会性形質の頻度は、$\bar{A} = \frac{u_{01}}{u_{01} + u_{10}}$ (タイポ？)
- 政治体の分裂
	- 各タイムステップごとに、各集団は、確率 $P_{disint} = \delta_0 + \delta_N N - \delta_A \bar{A}$ で分裂する。

##### シミュレーションの結果
Turchin  らは、このモデルを使って、紀元前1500年から紀元後1500年の間の、ユーラシア大陸の領土のダイナミクスをシミュレートした。

- 初期状態として、ユーラシア大陸中央部のステップ地帯との境界にあるすべてのグリッドは、軍事技術を持つとする。
	- これは、この地域の遊牧騎馬民族が、このモデルで想定している軍事技術の創始者であることと対応している。
- 現実に起こった巨大な政治体の時空間分布を再現することが、彼らの目的であった。
- シミュレーションの結果、社会性形質の分布は、現実の大規模な集団分布と類似していた。
	- e.g. まず、エジプト、メソポタミア、中国に巨大な政治体が出現し、その後その周辺地域に広がっていった
    - ⇒政治体が位置する地理的条件が、政治体の規模の拡大をもたらしやすい
- シミュレーションの条件を変えて、どのような要因が現実と類似した巨大政治体の時空間分布を作り出しているのかを検討
	- 1要因＝軍事技術の発信地
    	- 軍事技術の発信地がステップ地帯という初期条件をかえると、異なった巨大政治体の時空間分布になった
- このモデルでは、軍事技術と社会性形質の伝達経路が異なっていたが、このように文化形質によって伝達経路が異なると考えるのは自然なことである。

### 5.3 集団間の文化的距離
集団間の文化多様性の定量化をしたい。そのために集団間の文化的距離を数値化について考える。
#### 5.3.1 Hamming 距離と Jaccard 距離
２値ベクトル $a \in \{0,1\}^{n_a}$ が文化形質のコード化として最も頻繁に用いられる。
各成分　$a_i\in\{0,1\}$ は対応する文化形質の有無を表す。

##### $ \{0,1\}^{n_a}$ 上の距離
###### Hamming distance(ハミング距離)
一般的な配列で、二つの配列のうち値が異なる要素の個数。

二値ベクトルの場合、$ \{0,1\}$ を、体 $\mathbb{F}_2$ として見て、距離 $d_H:\mathbb{F}_2^{n_a} \times \mathbb{F}_2^{n_a} \rightarrow \mathbb{R}$ を
$$d_H(a, b) := |a+b|$$
として書ける。ここで $a \in \mathbb{F}_2^{n_a}$ に対し、$|a| := \sum_{i=1}^{n_a} |a_i|, (|\cdot|:\mathbb{F}_2 \hookrightarrow \mathbb{R})$。また、$a+b$ はベクトルとしての和。
###### Jaccard distance
二つの集合 $A,B$ 同士の、差異の大きさを測る、値。 Jaccard index の補数。
$$d_J(A, B) := \frac{|A \bigtriangleup B|}{|A \cup B|} = 1 - \frac{|A \cap B|}{|A \cup B|} = 1 - J(A, B)$$
ここで、$A \bigtriangleup B = (A \cup B) - (A \cap B)$ は symmetric difference(対象差)、 $J(A, B) = \frac{|A \cap B|}{|A \cup B|} $ は、Jaccard index（ジャカード指数）と呼ばれ、二つの集合の重なりの度合いを表す。画像処理では Intersection over Union = IoU という名称で、物体検出、セグメンテーションなどの精度評価に使われている。

ちなみに、二値ベクトル $a,b \in \mathbb{F}_2^{n_a}$ の場合は、
$$J(a,b)= \frac{|a \odot b|}{n_a}$$
で得られる。$\odot$ は成分ごとの積( Hadamard product)。
#####  距離を用いた研究例
 Deborah Rogers と Paul Ehrlich は、ポリネシアの 10 この島とフィジーのカヌーを対象とした研究にジャッカード距離を用いた。
 
 各集団に対して、カヌーのデザインの機能的要素を 96個、装飾的要素を38個用意して各集団間の、機能的要素と装飾的要素のジャッカード距離を計算し、これらの要素の間の相関は弱いことを示し、これらの要素が独立して変化していることを示した。

 機能的要素は装飾的要素よりもゆっくり変化する傾向があった。
これから、カヌーの機能的要素は、負の選択にさらされていると考えた。

#### 5.3.2 文化多様性の測度
文化の多様性の定量化のしかた。
#####  Heterozygosity(ヘテロ接合度)
文化形質 $i$ の頻度を $p_i$ として
$$ H = 1 - \sum_{i} p_i^2$$
で定義される。

用語
- 遺伝子座（locus）：特定の遺伝子または遺伝子マーカーが存在する染色体上の特定の固定位置（https://ja.wikipedia.org/wiki/%E9%81%BA%E4%BC%9D%E5%AD%90%E5%BA%A7 )
- 対立遺伝子(allel)：対立形質を規定する個々の遺伝子(https://ja.wikipedia.org/wiki/%E5%AF%BE%E7%AB%8B%E9%81%BA%E4%BC%9D%E5%AD%90)
- 接合性(zygosity)：生物の形質に関するアレルの類似性（https://ja.wikipedia.org/wiki/%E6%8E%A5%E5%90%88%E6%80%A7 ）

二倍体生物のある遺伝子座に、対立遺伝子（アリル）が $m$ 個あるとすると集団内での、各アリル $i$ の頻度を $p_i$ で書くと
$$\sum_i^m p_i = 1$$
だから、ヘテロ接合度は
$$ H =  1 - \sum_{i}^m p_i^2 = \left( \sum_j^m p_j \right)\left( \sum_k^m p_k \right) - \sum_{i}^m p_i^2 = \sum_{j \ne k} p_j p_k$$
と表せる。この形だと、この指標がヘテロ接合の多さ（異なるアリルを持った個体の頻度）を表す指標であることが分かりやすい。

集団中に、ある一つのアリル $i$ だけが存在する場合、すなわち
$$ p_j = \begin{cases}
1 & \quad j=i\\
0 & \quad \text{otherwise}
\end{cases}$$
の場合、ヘテロ接合度は $H = 0$ になる。

##### 固定指数
$$ F_{ST} := \frac{H_T - H_S}{H_T} $$
ここで、$H_T$ は全集団のヘテロ接合度、$H_S$ は各分集団のヘテロ接合度の平均。

Adrian Bell らは世界各地の集団の遺伝文化度と文化分化度を $F_ST$ を用いて測定した結果、遺伝子よりも文化の方が分化していた。この結果から彼らは、文化の集団間の分化が大きく、群淘汰が働く余地があるとした。（群淘汰はどういう意味？）
##### Shannon index
他にも、多様性の尺度して、 以下に定義される、Shannon index(シャノン指数)という指標もある。
$$-\sum_{i} p_i \ln{p_i}$$

### 5.4 集団構造と文化大進化プロセスの推定
多くの場合、文化大進化のプロセスの直接的な証拠は得られない。

よって、データから進化パターンを推定する際、間接的な証拠が用いられる。

- 言語は垂直伝達されると仮定する場合が多い
	- この場合、近い言語を話す集団が同じ文化を共有してる場合、それは垂直伝達の結果である可能性が高い。
- 地理的に隣接している集団間では水平伝達が起こりやすいと仮定されることが多い
	- 現代の通信技術が発達したのは、人類史のほんの最近の時代。
    - この場合、地理的に近い集団が同じ文化を共有しているとき、それは水平伝達の結果である可能性が高い。
- 文化形質は地理的に一様に分布しているわけではない。こうした文化形質における、地理的な偏りを**集団構造**と呼ぶ。

#### Carmela Guglielmino らによる研究(1995)
- Ethnographic Atlas より277個のアフリカの集団と、それらがもつ47個の文化形質を分析した。
- 47の文化形質は、以下のカテゴリーに分類されている
1. 親族構造
2. 経済
3. 社会階層
4. 性的分業
5. 家
6. その他
- 各文化カテゴリーの形質について、垂直伝達、水平伝達、環境適応の3つのモデルのどれが最もよく分布を説明できるかを検討した
	- 垂直伝達、あるいは人口拡散される文化形質は、文化進化速度が小さく、言語を共有している集団内で共有される。
    - 環境適応は、生態学的環境との相関で測られる
    - 水平伝達、あるいは文化伝搬があれば、地理的な拡散が見られるはず。
    	- 水平伝達の指標 $r = \frac{d}{d_S}$ ： $d=$ ある社会から最も近い社会までの距離、$d_S=$ その社会と同じ文化形質を共有している最も近い社会までの距離。
- 解析の結果
	- 親族構造は垂直伝達する傾向があった
    - 性的分業、その他の文化形質は水平伝達する傾向があった
    - 環境適応が示唆されるカテゴリーはなかった
    - さらに詳しく見た結果、生態的要因と相関している文化形質のほとんどは、言語とも相関していた。
    	- このことから、 Guglielmino らは、集団が拡散する際に、同じ環境に拡散する傾向があったのではないかと考えている
        - Borgerhoff Mulder は垂直伝達・水平伝達と環境適応は必ずしも相互に排他的なものではないことに注意を促している。
        	- 文化伝達によって、適応的な文化を獲得することも多い

#### 2010年以降盛んになった研究
$\Phi_{ST}$ という（固定指数？）集団遺伝学の指標を文化データに適用する研究が盛んになった。

このような困難がある事を理解しつつ、データと解析結果を慎重に吟味していくことが必要。

### 5.5 文化系統学

Darwin は進化を「変化を伴う由来（descent with modification）」と定義した。

変化を伴う由来（descent with modification）とは、共通の祖先から発生した個体が時間の経過とともに形質が変化し別の種になることである。

文化進化も、この変化を伴う由来ならば、多くの文化形質は継承されたものであり、それは必然的に系統を構成する。

生物の進化史を推定する**系統学**(phylogenetics) の方法を、文化現象に適用したのが**文化系統学**(cultural phylogenetics)である。

文化系統学によって文化間あるいは集団間の関係性を可視化することができる。

系統図は、ノード(node)とそれらをつなぐ枝(branch)からなる。

ノードのうち、現在のものを、葉ノード(leaf node)、あるいは操作的分類単位(operational taxonomic unit; OTU)という。

過去のもの(２つの枝が交わるノード)は、祖先ノード(ancsestral node)という。

#### 5.5.1 考古遺物の系統解析

文化系統学では対象間の類縁関係を復元する。

その対象も、系統推定の方法も様々。

ここでは、石器の系統関係を**最節約法**(maximum parsimony method) と呼ばれる手法を使って復元した研究を紹介している。

最節約法は
- 形質の変化は起こりにくいことを仮定し
- 「最も節約的」。つまり最も形質の変化が少ない系統樹を選択する方法
    - この「最も節約的」という原理は、生物進化において基本的な原理の一つ。文化進化でも同様。

Example:
最節約法について以下の図で説明する。

![pic3](chap5-mutation-example.png)

文化形質aをもつ共通祖先から、４つの集団が生まれたとする。Group1,2が文化形質Aをもち、Group3,4が文化形質aをもつとする。

このパターンを説明する時、左の図であれば、aからAへの変異が１回だけ起こったとすればよいが、右の図を使うと、２回変異が起こったと考えなければならない。

最節約な考え方に基づけば、左の図のほうがより良い系統樹ということが言える。

一般に、最節約な系統樹を求めるのは大変であるが、効率的なアルゴリズムが提案されている。

注意すべきは、最節約法に従ったからと言って、常に「正しい」樹形が得られるとはかぎらないということである。

しかし、直接観察できない過去を推定するには、何らかの基準が必要である。

最節約的原理は、文化進化研究において、これまでも、これからも、有効な視点で有り続けるであろう。

#### 5.5.2 言語系統学

文化系統学が最も成功を収めたのは、言語学の分野である。

言語系統樹を作る上でよく使われるのは、基礎語彙と呼ばれる特定の語彙である。

ある語が使われていれば1、使われていなければ0をコードして、01の配列を作り、そこから系統樹を推定する。

※同根語(cognate)は同じ単語として扱われる。

言語系統学では、類縁関係だけでなく、分岐年代が興味の対象となる。

分岐年代を推定することで、人類の移動に関する様々な仮説に答えることができる。

##### インド・ヨーロッパ語族の分岐年代の推定

例えば、 Russell Gray et al はインド・ヨーロッパ語族の分岐年代から、農耕・言語の共拡散仮説の検証を行っている。

農耕・言語共拡散仮説とは、農耕が発明されたあと、農耕民が在来の狩猟採集民を置き換えながら拡散していくことで、言語と農耕がともに分布を広げていくとする仮説である。

彼らは、インド・ヨーロッパ語族の起源について、2つの仮説を検討している

1. クルガン騎馬民族仮説：現在のロシア南部（あるいは黒海周辺）に住んでいたクルガン騎馬民族が、その拡散とともにインド・ヨーロッパ語族を拡散させた。
2. アナトリア仮説：約1万年前のトルコアナトリア周辺の農耕民の拡散が、インド・ヨーロッパ語族を拡散させた。

Gray らはインド・ヨーロッパ語族の分岐年代を約1万年前と推定し、アナトリア仮説が確からしいと結論づけている。

この結果は、2012年に行われた、データ量を増やした解析でも支持されている。

https://en.wikipedia.org/wiki/Galton%27s_problem#:~:text=Galton's%20problem%2C%20named%20after%20Sir,to%20experimental%20design%20as%20well.#### 5.5.3 文化の「系統樹らしさ」

言語系統がくをはじめとする文化系統学には批判も多い。

中でもよく問題にされるのが、水平伝達の影響である。系統学は基本的に垂直伝達しか起こらないと仮定している。

水平伝達が増えると、文化大進化の様相が、樹形から離れていく。

こうした批判は、文化進化のあり様を樹形で捉えるべきなのかどうかをめぐる論争と言えるだろう。

この問題に取り組んだ研究： Simon Greenhill et alによるシミュレーションがある。

彼らは水平伝達は、類縁関係の推定にはさほど大きな影響を与えないが、分岐年代は若く見積もられることを示した。

進化の様相が樹形的かどうかを検討する別の方法：系統樹の「木らしさ」を測る指標を用いる。

**保持指数**(retention index; RI)

$$ RI = \frac{最大変化数 - 観察された変化数}{最大変化数 - 最小変化数} $$

コラード、シェナン、テヘラニは、様々な文化形質と生物学的形質のRIを測定。その結果、文化的/生物学的形質の平均の保持指数に大きな違いは見られなかった。

つまり、少なくともデータ上は文化進化だからといって、生物進化よりも樹形的でないとは言えないことが示唆される。

しかし、特に十分な水平伝達がある場合の保持指数の性質は理論的に十分検討されているとは言い難い。

とたえば、クレーマらは、シミュレーションによって、水平伝達が低い場合保持指数は高くなる傾向があるが、不確実性も高く、保持指数が低いことが直接的に水平伝達の高さを意味するわけではないことを報告している。

### 5.6 祖先形質復元
5.5 では、文化系統学の研究として、文化の系統関係や分岐年代はを明らかにする試みについて紹介してきた。


他に、文化系統樹からわかることとして、形質の**共進化**(coevolution)パターンを抽出することもできる。

共進化とは、2つ以上の生物種あるいは形質が、相互依存的に進化することである。

祖先ノードの形質を遡って復元していく**祖先形質復元**(ancestral state reconstruction)と呼ばれる手法で、こうした共進化パターンが解析できる。

#### 5.6.1 牛は母系の敵：ゴルトン問題

特定の2つの文化形質がいくつもの文化で見られる場合、その組み合わせに意味があるのかないのか、もしくは単に祖先が共通だからなのかは、すぐにはわからない。

これは文化に関するデータが統計的に独立でないために生じる問題であり、**ゴルトン問題**(Galton's problem)と呼ばれている。

[Galton's problem(wikipedia)](https://en.wikipedia.org/wiki/Galton%27s_problem#:~:text=Galton's%20problem%2C%20named%20after%20Sir,to%20experimental%20design%20as%20well.)

ここでは、ゴルトン問題の一例として、家畜(ウシ, cattle)所有と父系制(patrilineality)の共進化の研究を取り上げている。

多くの集団において、各人がもつ財産やアイデンティティは、血縁関係を通して継承される。

父方の祖先から相続される場合を**父系**(patrilineal)、母方から相続される場合を**母系**(matrilineal)とよぶ。

人類学では、古くから、「ウシは母系の敵」と言われきた。

David Aberle によれば、84の母系社会のうち、47が園耕農業(horticulture)を営んでいるが、ウシを所有しているのはそのうちの19の社会のみである。

また、農園を営む188の社会のうち、47が母系であり、242の牧畜社会のうち母系は19だけである。

この事実を指して、Aberle は「ウシは母系の敵」と表現したが、これはプロセスではなく、パターンについての言及である。

どのようなプロセスによってこのようなパターンが生まれたのかはこの段階ではわからない。

Aberle はカイ二乗検定を使って、文化要素間の相関、つまりある文化要素Aを持っている集団が別の文化要素Bを持っていやすいかを調べた。

しかし、文化データは、個体同士が独立であるとみなせないため、このような統計検定から単純に結論を導くことはできない。

分岐前の集団が、父系かつ家畜所有であった場合、その娘集団どうしも、突然変異がない限り、同じ2つの形質を引き継ぐ。

よって、ある2つの集団が、同じ形質の組み合わせを共有していたとしても、それが、同じ祖先からの継承によるものなのか、生態学的・社会的な要因によるものなのかの区別はつかない。

すなわち、統計学的には、これらの形質の間の関係が独立でないことを意味し、データの独立性を仮定する多くの統計検定はそのままでは使えない。

進化生物学者たちは、種間の系統関係の効果をコントロールしつつ、形質と様々な要因との関係を調べる方法として、 **系統比較法**(phylogenetic comparative method)を構築してきた。

Clare Holden と Ruth Mace は、この系統比較法を応用することで「ウシは母系の敵」というパターンが本当にあるのか、それとも、単に継承によるものなのかを検討した。

[Spread of cattle led to the loss of matrilineal descent in Africa: a coevolutionary analysis](http://www.ceacb.ucl.ac.uk/cultureclub/files/2004-2005/CC2005-04-21-holden_Mace_2003.pdf)

- まず、Bantu語族の言語系統樹を推定した。
- その後、 Murdock の Ethnographic Atlas から、その集団の相続形態と家畜所有の有無のデータを得た。
- そして、現代の集団の形質から、過去の集団の形質を推定していく。
    - 例えば、2つの集団AとBが、ともに母系であれば、その共通祖先もまた母系である可能性が高いであろう。
    - この様に、現代の集団から、全集団の共通祖先まで遡って形質の状態を復元する。
- 最も古い祖先から、形質の状態がどのように遷移していったのかを推定する。
    - ここでは、着目している形質の状態は、(母系・家畜なし)、(母系・家畜あり)、(父系・家畜なし)、(父系・家畜あり)の4つである。
    - どの状態からどの状態への遷移が起こりやすいかを調べた。
- 結果
    - (母系・家畜あり)→(母系・家畜なし)or(父系・家畜あり)の変化が起こりやすかった。
    - 一方、(父系・家畜あり)からの変化は起こりにくかった。
- 彼らはこの社会変化パターンを、婚資の観点から解釈している。
    - Bantu語族の社会では、結婚に際し、男性が相手の家族に家畜を送る習慣がある。
    - そのため、財産を男性に相続することは、息子の結婚できる可能性を上げるための有効な戦略だったのではないかとしている。

#### 5.6.2 政治形態の複雑化
祖先形質復元の別の例として、 Thomas Currie らによる政治形態の複雑化の研究がある。

- 古典的な社会進化論によれば、あらゆる社会は、画一的な発展段階のどこかに位置づけられ、やがて次の段階に遷移していくと考えられていた。
    - この主張は、現代の文化人類学・社会人類学では退けられている。
    - しかし、実際に複雑さがどの様に進化するのかはいまだ明らかではない。
- 社会の複雑性のダイナミクスを考える際の重要な一要素が、政治形態の複雑性である。
- Currie らは社会の「複雑さ」に関する仮説を、定量的に検証することを試みた。
    - オーストロネシア語族の集団を対象とした。(Ethnographic Atlas)
    - 政治形態の複雑さの指標として、意思決定の階層の数に着目。
        - 各社会を「首長なし」「単純な首長制」「複雑な首長制」「国家」の4段階に分類した。（右に行くほど複雑）
    - まず、オーストロネシア語族の系統樹を推定し、対象集団の歴史的類縁関係を復元
        - 既知の政治形態の情報に基づき、系統樹の各時点での政治形態の状態を推定
        - これらの復元情報から、政治形態の遷移規則を計算。
    - 得られた遷移規則が、過去に提唱された社会進化の仮説のうちどれと最も整合的かを、モデル選択を用いて検討した
    - 明らかになったのは2つの傾向だった。
        - (1) 複雑さの増加は段階的に起こる
            - 首長なし→国家のような変化は起こりにくい
        - (2) 複雑さが急に減少する変化は起こりうる
            - 国家→複雑な首長制or首長なし両方とも起こりうる
- この研究に対する反省点
    - 対象がオーストロネシア語族のみ
    - 政治形態の複雑さの指標を一つしか使っていない
        - 解析の直接的な結果と、そこから間接的に示唆されることを区別しなければならない。
    - この研究が明らかにしたのは、プロセスではなくパターンである。
        - 乱暴に言うと、遷移状態の方向と数をようやくしただけ（ある状態遷移を引き起こす原因を特定していない）。
        - しかし、異なるプロセスが同様なパターンを生むことがある
            - たとえば、階層の減少は、自然環境悪化による人口減少によったり、支配階層への不満によるのか、など
    - 祖先形質復元においても、水平伝達の問題はつきまとう。
        - Mace らは「水平伝達は操作上は突然変異として扱える」と述べている

### 5.7 幾何学的形態測定学の文化への応用

ここまで取り上げてきた研究では、文化のデータ化は、その分野の研究者によってなされているが問題点がある。
1. データが持っていた情報が、分類時に失われる恐れがある
2. 他の研究者と分類基準の共有が難しい

などである。

ここでは、客観的なデータ化基準を定め易い対象の研究として、**幾何学的形態測定学**(geometric morphometrics)を紹介している。

**形態測定学**(morphometrics)は、対象（主として生物）の形を定量的に計測し、解析するための方法群である。

「幾何学的」とついているのは、「伝統的」形態測定学と対比するため。

- 伝統的手法：「計測点」を定め、それらの距離や角度を測定し、形状を定量化。
    - 人間が特徴的と感じる、少ない数の点を恣意的に選択する。
    - その過程で、細かい幾何学的情報が失われる
- 幾何学的：対象の座標をそのまま解析。
    - 簡単に言うと、計測点を増やした。
    - 人間のバイアスが少なくなった
    
幾何学的形態測定学では、「形態」と「形状」は明確に区別される。
- **形態**：サイズ-シェイプ (size-shape, form)などとも呼ばれ、回転、平行移動に対して不変な幾何学的情報のことである。（invariant under 2D rigid motions）
- **形状**(shape):回転、平行移動、拡大収縮に対して不変な幾何学的情報のことである。（invariant under 2D affine transformations）

幾何学的形態測定学を以下の２つに大別できる
1. 標識点ベース形態測定学(landmark-based morphometrics)
2. 輪郭ベース形態測定学(outline-based morphometrics)

#### 5.7.1 標識点ベース形態測定学
**標識点**(landmark)とは、解析対象間で対応する形態的な特徴点のことである。

標識点ベース形態測定学では、標識点の集合により形を記述する。

標識点はすべての対象で共通していなければならない。

##### 標識点の分類
標識点の分類の仕方にはいくつかあるが、例えば以下のものがある

- 科学的標識点(scientific landmark):何らかの専門的知見に基づき配置された点
    - 例:土器の特定箇所のくびれ、製作時の接合点
- 数学的標識点(mathematical landmark): 対象の数学的・幾何学的特徴を基準に配置された点
    - 例: 曲率最大の点、端点
- 疑標識点(pseudo-landmark):科学的・数学的標識点を参照点とし、単純な規則にしたがって配置された点
    - 輪郭に沿って等間隔に配置された点
    
別の分類(Fred Bookstein)

- Type I:形態構造の接する点
    - ex. 前方後円墳の前方部と後円部の接合部、刀剣の刃と柄が接する点
- Type II:最大曲率点
    - 鉄鏃の先端、土器の胴部の曲率最大点
- Type III:その他、幾何学的に定義可能な点
    - 端点、図心、線分の交点
    
##### 位置合わせとサイズの規格化(プロクラステス再配置)
標識点ベース形態測定学の第一歩は位置合わせとサイズの規格化である。

最もよく使われる方法の一つは、重心を合わせる方法。

**プロクラステス再配置**(Procrustes superimposition)と呼ばれる規格化を、二次元に投影された銅鏃を使って説明している。

- 銅鏃に３つの標識点をとる。
- 標識点の重心を求める。
- すべての重心を原点に移動する。
- 各銅鏃の重心サイズを求め、それを使ってそれぞれの銅鏃を規格化する。
    - 重心サイズ:各標識点から重心までの距離の平方和の平方根
- 対応する標識点間の距離が最小になるように回転させる。

##### 多変量解析
主成分分析(PCA)がよく行われる（次元削減）。


#### 5.7.2 輪郭ベース形態測定学：楕円フーリエ解析
輪郭ベース形態測定学は、対象の輪郭を解析する。輪郭とは、対象とそれ以外を分ける境界を意味する。関数で対象の輪郭を近似する。

楕円フーリエ解析は、輪郭ベース形態測定学の代表的な手法の一つ。

輪郭をパラメータ $t$ の周期関数 $ [0,T] \ni t \mapsto (x(t), y(t)) \in \mathbb{R}^2$ $(x(0), y(0)) = (x(T), y(T))$ として表し、各座標関数をフーリエ級数展開して解析するものである。$x(t), y(t)$ がそれぞれある一定の条件を満たせば、

$$
\begin{eqnarray}
x(t) &=& \frac{a_0}{2} + \sum_{i=1}^{\infty} a_i \cos{(i \omega x)} + \sum_{i=1}^{\infty} b_i \sin{(i \omega x)}\\
y(t) &=& \frac{c_0}{2} + \sum_{i=1}^{\infty} c_i \cos{(i \omega x)} + \sum_{i=1}^{\infty} d_i \sin{(i \omega x)}
\end{eqnarray}
$$

と書ける。通常、$a_0,c_0$ は省略される。サイズ規格化のための楕円の長軸の長さで割る方法や、面積で割る方法がある。

フーリエ級数の係数 $a_i, b_i, c_i, d_i$ を特徴量とみなせるが、係数の数が多いため、主成分分析を用いて次元削減することが多い。

本では、メスリ山古墳と東大寺山古墳の銅鏃の輪郭を楕円フーリエ解析とPCAを用いて、２つの古墳の銅鏃がそれぞれ明確なクラスタを形成していることを示している。

前節で扱った、３つの特徴点を用いた標識点ベースの解析では、メスリ山古墳の１個の銅鏃が飛び値として、東大寺山のクラスタに混ざっていたが、こちらの手法では分離できている。

これは、楕円フーリエ解析のほうが、扱う輪郭の情報量が多いことで可能になった。

「ただ、一概にすべてのケースで、輪郭ベースのほうが標識点ベースより優れているとはいえず、ケースバイケースで手法を使うのが良い。」とある。

この銅鏃の例で扱ったのは、あくまで、２つの古墳の銅鏃の形の傾向（パターン）の分析であり、「なぜそうなったのか」（プロセス）の分析はしていない。

今回の場合は、２つの古墳に時期的な差があることがわかっているので、時間に伴う変遷と言えるが、更に、なぜそのような変遷が起こったのかについて答えるには、幾何学的形態測定学や、文化進化の手法だけでは難しく、他の分野の手法が必要である。

楕円フーリエ解析を適用したその他の対象の例
- 弥生式土器（遠賀川式土器）の地理的変異のパターンを定量化
- 小森らによるシャンプーボトルの形状から人々が受ける印象の解析
- 発展として、3次元データを扱う球面調和関数を使った解析

##### 幾何学的形態測定学の使われ方
文化進化における幾何学的形態測定学の使われ方の例
- どのようなクラスタがいくつできるかをもとに、文化進化について情報を引き出す
    - 空間的・時間的形態変化の定量化
    - 機能、用途による形態差の定量化
- 形態のばらつき具合そのものを対象にする
    - 例:強大な権力の出現によって、遺物の規格性がたかまる→権力の出現前後でばらつき具合が変化するであろう
    
##### よくある誤解
幾何学的形態測定学では、基準を明示化するので、手法やデータの「客観性」が高いように思われ、「主観」から自由な「客観的」な解析ができるという誤解がある。

幾何学的形態測定学では、ある基準をすべての対象に適用することができるが、その基準の決定は、研究対象分野の専門知識から自由ではないので、人間による恣意性が必ず入ってくる。

定量化の手法が変われば、同じ対象でも得られる結果は変わってくる。それは、物を見る視点を変えるようなものである。

よって、絶対的な客観性というものは存在しない。

また、
- 採用した幾何学的基準が、対象物の製作者や使用者の意図と必ずしも一致しない
- 研究デザインとして、幾何学的形態測定学を使用したとしても、その他の研究分野の判断やカテゴライゼーションと整合的でなければならない
    - 最終的に結果を使用したい分野での概念から全くずれている、的はずれな基準で形態測定をしてしまうかもしれない
    
だが、**基準を明示することは、学術の蓄積に貢献する。手法に依存するのではなく、何ができて、何ができないのか、個別の研究で明らかにしたいことは何なのかを、具体的に整理することが必要。**
